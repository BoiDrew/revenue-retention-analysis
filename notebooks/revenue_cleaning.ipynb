```python
import pandas as pd
df = pd.read_excel("../data/raw/Online_Retail.xlsx")
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df[df['InvoiceNo'].str.startswith('C', na=False)]
```

```python
#Remove cancelled oreders or transcations

df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
```

```python
# Drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])
```

```python
# Create Revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
# Date formatting
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df.head()
```

```python
# Save cleaned data
df.to_csv("../data/cleaned/online_retail_cleaned.csv", index=False)
```

## Data Cleaning Decisions
- Removed cancelled transactions (negative quantity)
- Removed zero-priced items
- Dropped records with missing CustomerID
- Created Revenue metric
- Extracted InvoiceMonth for time-based analysis
